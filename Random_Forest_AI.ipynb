```python
#treating the outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## **Model building**

```python
#converting the cat into the numerical var
df=pd.get_dummies(df,columns=['ShelveLoc','Urban','US'],drop_first=True,dtype=int)
```

```python
df
```

```python
# importing the needed libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# importing the scaler
scale=StandardScaler()
```

```python
# splitting the data into the feature and the target
x=df.drop('Sales',axis=1) # Feature varabile
y=df['Sales'] # target variable
```

```python
# scaling the feature variable
x=scale.fit_transform(x)
```

```python
#splitting the data into the train test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# fitting the model
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
#predecting the data
y_pred=model.predict(x_test)
```

```python
#evaluting the metrices
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
#XG boosting
import xgboost as xgb
```

```python
x1=x.copy()
y1=y.copy()
```

```python
#splitting the data
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python

model = xgb.XGBRegressor(objective='reg:squarederror', # Objective for regression
                         n_estimators=100,          # Number of boosting rounds
                         learning_rate=0.1,         # Step size shrinkage used in update to prevent overfitting
                         max_depth=5,               # Maximum depth of a tree
                         random_state=42)           # Random seed for rep
```

```python
# Fit the model to the training data
model.fit(x_train, y_train)
```

```python

# Predict on the test data
y_pred = model.predict(x_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("XGBoost Model Evaluation:")
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
x2=x.copy()
y2=y.copy()
```

```python
x_test,x_train,y_test,y_train=train_test_split(x2,y2,test_size=0.2,random_state=42)
```

```python
# Gradient Boosting Regressor
gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbr_model.fit(x_train, y_train)
y_pred_gbr = gbr_model.predict(x_test)
```

```python
print("\nGradient Boosting Regressor Evaluation:")
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_gbr))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_gbr))
print("R-squared:", r2_score(y_test, y_pred_gbr))
```
