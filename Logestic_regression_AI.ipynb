```python
# splitting the data target and the feature
X=df.drop('loan',axis=1)
y=df['loan']
```

```python
#importing the pca(Principle compount analysis)
from sklearn.decomposition import PCA
```

```python
#fitting the pca
pca = PCA(n_components=8)
X_pca = pca.fit_transform(X)
```

```python
x=X_pca
```

```python
x
```

```python
# convertig the array into the data frame
x = pd.DataFrame(X_pca, columns=["pca1","pca2","pca3","pca4","pca5","pca6","pca7","pca8"])
display(x.head())
```

```python
# importing the standard scalar
from sklearn.preprocessing import StandardScaler
```

```python
# Scaling the data
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
# splitting the data into train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#loding the model
model = LogisticRegression()
```

```python
#fitting the model
model.fit(x_train,y_train)
```

```python
# predecting the data
y_pred = model.predict(x_test)
```

```python
# evaluting the data
from sklearn.metrics import accuracy_score,recall_score,precision_score,confusion_matrix,classification_report
```

```python
# predecting the accuracy
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```
